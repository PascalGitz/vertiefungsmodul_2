```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def M_y(F, x):
    return F*np.flip(x)
```

```python
x_array = np.linspace(0,5,100)
F_array = np.linspace(0,40,2)

M_y_list = []
for Force in F_array:
    M_y_array = M_y(Force, x_array)
    M_y_list.append(M_y_array)
    
```

```python
for i,M in enumerate(M_y_list):
    plt.plot(x_array, M, color='orange')
    plt.ylim(0,200)
    plt.title('Biegemomentenverlauf')
    plt.savefig(f'M_y_{i}.jpg')
    plt.show()
```

```python
M_for_chi = np.array([0,30,30,170,200])
chi_for_M = np.array([0,0.001,0.003,0.025,0.1])
```

```python
plt.plot(chi_for_M, M_for_chi, color='black')
plt.title('Momenten-Krmmungs-Beziehung')
plt.savefig('momenten_krummungsbez.svg')
plt.show()
```

```python
def chi_from_M(M):
    chi_for_x = np.interp(M, M_for_chi, chi_for_M)
    return chi_for_x

chi_x_list = []
for M in M_y_list:
    chi_x_list.append(chi_from_M(M))

for i,chi in enumerate(chi_x_list):

    plt.plot(x_array, chi, color='black')
    plt.ylim(0,0.1)
    plt.title('Krmmungsverlauf')
    plt.savefig(f'chi_x_{i}.jpg')
    plt.show()
```

```python
for i,chi in enumerate(chi_x_list):
    plt.plot(x_array,np.cumsum(np.cumsum(chi)), color='blue')
    plt.ylim(0,160)
    plt.gca().invert_yaxis()
    plt.title('Biegelinie')
    plt.savefig(f'w_{i}')
    plt.show()
```
